```python
import pandas as pd
import numpy as np
```

```python
clean= pd.read_csv('data_cleaning_challenge.csv')
clean.head(5)
```

```python
#dropping the last 2 columns
drop= clean.drop(columns=['Unnamed: 9','Unnamed: 10'])
drop.head(5)
```

```python
dropna= drop.loc[drop['Row Type'].notna()]
dropna.head(5)
```

```python
columns=[]                    #creating an empty column to append counter later
counter=0                
for i in dropna['Row Type']:  #selecting all the items in the row type column
    if "first name" in i:     #if first name is present in all the items in the row type columns count them 
        counter= counter+1
    columns.append(counter)   #add the counter in the list 

    
```

```python
dropna['Iteration']= columns 
```

```python
dropna.head(3)
```

```python
#USING THE STR FUNCTION
names= dropna.loc[dropna['Row Type'].str.contains('first name')]
names= names.rename(columns={'Row Type':'First Name', 'Iter Number':'Last Name', 'Power1':'Date'})
names= names.drop(columns=['Speed1', 'Speed2' ,'Electricity' ,'Effort','Weight' ,'Torque' ])
names['First Name']= names['First Name'].str[12:]
names['Last Name']= names['Last Name'].str[11:]
names['Date']= names['Date'].str[6:]
```

```python
names
```

```python
noname=dropna.loc[~dropna['Row Type'].str.contains('first name')]
noname
```

```python
#NOW WE WILL JOIN BOTH SHEETS BY JOIN METHOD
merged= pd.merge(left=names, right=noname, how='inner', on='Iteration')
merged
```

```python
merged.to_excel('Data Cleaning Practice Complete.xlsx', sheet_name='Result')
```
